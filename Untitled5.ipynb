```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import random

# Parameters
num_rows = 10000
start_date = datetime.strptime("2020-01-01", "%Y-%m-%d")
end_date = datetime.strptime("2024-12-31", "%Y-%m-%d")
date_range = pd.date_range(start=start_date, end=end_date, freq="D")
dates = np.random.choice(date_range, size=num_rows)

# Categories
expense_categories = ["Groceries", "Rent", "Transport", "Utilities", "Entertainment", "Healthcare", "Shopping", "Education"]
income_sources = ["Salary", "Freelance", "Investment Return"]
all_categories = expense_categories + income_sources
probabilities = [0.10]*8 + [0.08, 0.06, 0.06]  # sum = 1

# Amount generator
def generate_amount(date, category):
    month = date.month
    if category in income_sources:
        base = 2000 if category == "Salary" else 500
        return round(max(0, base + np.random.normal(0, 100)), 2)
    else:
        if category == "Rent":
            return round(1000 + np.random.normal(0, 50), 2)
        elif category == "Groceries":
            return round(200 + month * 2 + np.random.normal(0, 30), 2)
        elif category == "Entertainment" and month == 12:
            return round(300 + np.random.normal(0, 50), 2)
        else:
            return round(random.uniform(10, 500), 2)

# Generate dataset
categories = np.random.choice(all_categories, size=num_rows, p=probabilities)
amounts = [generate_amount(pd.to_datetime(d), cat) for d, cat in zip(dates, categories)]

# Create DataFrame
df = pd.DataFrame({
    "Date": pd.to_datetime(dates),
    "Amount": amounts,
    "Category": categories
}).sort_values("Date")

# Save to CSV
df.to_csv("structured_finance_data.csv", index=False)

# Download
from google.colab import files
files.download("structured_finance_data.csv")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
from google.colab import files

# Upload CSV file
uploaded = files.upload()
filename = next(iter(uploaded))
print(f"Uploaded file: {filename}")

# Load data
df = pd.read_csv(filename)
print(df.head())

# Ensure required columns
assert {'Date', 'Amount', 'Category'}.issubset(df.columns), "CSV must have 'Date', 'Amount', 'Category' columns"

df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values('Date')

# -- Category-wise Linear Regression Forecast Function --

def predict_by_category(df, months_to_predict=3):
    forecast = {}
    for cat in df['Category'].unique():
        cat_df = df[df['Category'] == cat].copy()
        cat_df['Month'] = cat_df['Date'].dt.to_period('M')
        monthly = cat_df.groupby('Month')['Amount'].sum().reset_index()
        monthly['MonthNum'] = np.arange(len(monthly))

        X = monthly['MonthNum'].values.reshape(-1, 1)
        y = monthly['Amount'].values

        model = LinearRegression()
        model.fit(X, y)

        future_months = np.arange(len(monthly), len(monthly) + months_to_predict).reshape(-1, 1)
        preds = model.predict(future_months)
        preds = np.maximum(preds, 0)  # No negative spending

        forecast[cat] = [round(float(p), 2) for p in preds]
    return forecast

# -- LSTM Time Series Forecast Function --

def lstm_forecast(data, n_steps=3, epochs=50):
    data = data.reshape(-1, 1)
    generator = TimeseriesGenerator(data, data, length=5, batch_size=1)
    model = Sequential([
        LSTM(50, activation='relu', input_shape=(5, 1)),
        Dense(1)
    ])
    model.compile(optimizer='adam', loss='mse')
    model.fit(generator, epochs=epochs, verbose=0)

    # Forecast next n_steps
    preds = []
    batch = data[-5:].reshape((1, 5, 1))
    for _ in range(n_steps):
        pred = model.predict(batch)[0][0]
        preds.append(pred)
        batch = np.append(batch[:,1:,:], [[[pred]]], axis=1)
    preds = np.maximum(preds, 0)
    return preds

# Run category-wise forecast
category_forecast = predict_by_category(df)
print("Category-wise forecast (next 3 months):")
print(category_forecast)

# Prepare data for LSTM forecast: monthly total expenses
monthly_total = df.groupby(pd.Grouper(key='Date', freq='M'))['Amount'].sum().values
lstm_preds = lstm_forecast(monthly_total)
print("LSTM forecast (next 3 months):")
print([round(float(p), 2) for p in lstm_preds])

# Visualize category-wise example (first category)
first_cat = list(category_forecast.keys())[0]
cat_df = df[df['Category'] == first_cat].copy()
monthly_cat = cat_df.groupby(pd.Grouper(key='Date', freq='M'))['Amount'].sum()

plt.figure(figsize=(8,4))
plt.plot(monthly_cat.index, monthly_cat.values, label='Actual')
future_dates = pd.date_range(monthly_cat.index[-1] + pd.offsets.MonthBegin(), periods=3, freq='M')
plt.plot(future_dates, category_forecast[first_cat], 'r--', label='Forecast')
plt.title(f"{first_cat} Spending Forecast")
plt.legend()
plt.show()

# Visualize LSTM forecast
monthly_dates = pd.date_range(df['Date'].min(), periods=len(monthly_total), freq='M')
plt.figure(figsize=(8,4))
plt.plot(monthly_dates, monthly_total, label='Actual')
future_dates = pd.date_range(monthly_dates[-1] + pd.offsets.MonthBegin(), periods=3, freq='M')
plt.plot(future_dates, lstm_preds, 'r--', label='LSTM Forecast')
plt.title("LSTM Total Spending Forecast")
plt.legend()
plt.show()
```

```python
# 📌 STEP 1: Install dependencies
!pip install -q matplotlib pandas tensorflow

# 📌 STEP 2: Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from google.colab import files

# 📌 STEP 3: Upload CSV file
print("Upload your expense CSV (columns: Date, Amount)")
uploaded = files.upload()
file_name = list(uploaded.keys())[0]

# 📌 STEP 4: Load and preprocess data
df = pd.read_csv(file_name)
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.to_period('M')
monthly = df.groupby('Month')['Amount'].sum().reset_index()
monthly['Month'] = monthly['Month'].astype(str)
monthly['Month_Num'] = range(len(monthly))

X = monthly[['Month_Num']].values
y = monthly['Amount'].values

# 📌 STEP 5: Build simple linear regression model using Keras
model = Sequential([
    Dense(1, input_shape=(1,), activation='linear')
])

model.compile(optimizer='adam', loss='mse')

# 📌 STEP 6: Train model
model.fit(X, y, epochs=200, verbose=0)

# 📌 STEP 7: Save model as .h5
model.save('linear_model.h5')
print("Model saved as linear_model.h5")

# 📌 STEP 8: Predict next 3 months
future_months = np.array([len(monthly), len(monthly)+1, len(monthly)+2]).reshape(-1, 1)
future_preds = model.predict(future_months)

# 📌 STEP 9: Plot past and future expenses
plt.figure(figsize=(10, 5))
plt.plot(monthly['Month_Num'], y, marker='o', label='Past Expenses')
plt.plot(future_months, future_preds, marker='x', linestyle='--', label='Forecast')
plt.xlabel("Month Index")
plt.ylabel("Total Spent")
plt.title("Monthly Spending Forecast (Keras Linear Model)")
plt.legend()
plt.grid(True)
plt.show()

# 📌 STEP 10: Show predictions
for i, val in enumerate(future_preds, start=1):
    print(f"📅 Month+{i}: ${val[0]:.2f}")

# 📌 STEP 11: Download saved model (optional)
from google.colab import files
files.download('linear_model.h5')
```
